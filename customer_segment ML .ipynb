```python
features = [
    "Income", "Age", "Recency", "Tenure_days", "Total_Spend", "NumWebVisitsMonth", "NumWebPurchases","NumStorePurchases"
]
```

```python
X = df[features].copy()
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(X)
```

```python
X_scaler
```

```python
wcss = []

for i in range(1,11):
    kmean = KMeans(n_clusters=i)
    kmean.fit(X_scaler)
    wcss.append(kmean.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,marker = 'o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS (Inertia)')
plt.grid(True)
plt.show()
```

```python
kmeans_final = KMeans(n_clusters=6)
df["Cluster_KMeans"]= kmeans_final.fit_predict(X_scaler)
df["Cluster_KMeans"] 
```

```python
cluster_profile = df.groupby("Cluster_KMeans")[features].mean().round(2)
cluster_profile
```

```python
df["Cluster_KMeans"].value_counts()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaler)
df['PCA1'],df['PCA2'] = X_pca[:,0],X_pca[:,1]
```

```python
X_pca
```

```python
sns.scatterplot(x = 'PCA1',y = 'PCA2',data = df,hue ='Cluster_KMeans',palette = 'Set1')
plt.title("Customer Segmentation (PCA)")
```
